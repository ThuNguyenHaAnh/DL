```python
import os
import numpy as np
import pickle
from keras.api.preprocessing import image
from keras.api.models import Model
import tensorflow as tf
from keras.api.applications.resnet50 import preprocess_input
from keras.api.models import load_model
from google.colab import drive
from tqdm import tqdm
# Load m√¥ h√¨nh t·ª´ Google Drive ho·∫∑c th∆∞ m·ª•c /content/
drive.mount("/content/drive", force_remount=True)
model_path = "/content/drive/My Drive/Model/resnet50_2.h5"  # Ho·∫∑c "/content/inceptionv3_best3.h5"
model = load_model(model_path)

# L·∫•y feature extractor
feature_model = Model(inputs=model.input, outputs=model.get_layer("global_average_pooling2d").output)

class_names = ["Charts", "Maps", "Others", "Photographs", "System_Model"]

def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

def extract_features(img_path):
    img_array = preprocess_image(img_path)
    features = feature_model.predict(img_array)
    return features

def save_features(data_folder, output_file):
    feature_dict = {class_name: [] for class_name in class_names}

    for class_name in class_names:
        class_path = os.path.join(data_folder, class_name)
        if not os.path.isdir(class_path):
            continue

        print(f"üîç ƒêang x·ª≠ l√Ω l·ªõp: {class_name}")
        for img_name in tqdm(os.listdir(class_path), desc=class_name):
            if img_name.lower().endswith((".jpg", ".png", ".jpeg")):
                img_path = os.path.join(class_path, img_name)
                feature_vector = extract_features(img_path)
                feature_dict[class_name].append((img_name, feature_vector))

    # L∆∞u d·ªØ li·ªáu v√†o file pickle
    with open(output_file, "wb") as f:
        pickle.dump(feature_dict, f)

    print(f"‚úÖ ƒê√£ l∆∞u ƒë·∫∑c tr∆∞ng v√†o {output_file}")

# Ch·∫°y l∆∞u vector ƒë·∫∑c tr∆∞ng
data_folder = "/content/drive/My Drive/data"  # Ho·∫∑c "/content/data"
output_file = "/content/drive/My Drive/Report/features_50_1.pkl"  # Ho·∫∑c "/content/features.pkl"
save_features(data_folder, output_file)
```
